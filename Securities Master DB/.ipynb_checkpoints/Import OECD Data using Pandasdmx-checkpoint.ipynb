### Connecting to an SDMX web service, caching

First, we instantiate pandasdmx.api.Request. The constructor accepts an optional agency ID as string. The list of supported agencies can be viewed here, or as shown below.

```python
from pandasdmx import Request
```

#### Note: you can view the available agencies with 'Request.list_agencies()'

```python
oecd = Request('OECD')
data_response = oecd.data(resource_id='MEI_CLI', key='all?startTime=2018') #You can glean the resource id from the url of the page where the data zip is downloaded
df = data_response.write(data_response.data.series, parse_time=False)
#df.to_csv('c:\\Temp\\test_lei.txt', sep='\t')
```
