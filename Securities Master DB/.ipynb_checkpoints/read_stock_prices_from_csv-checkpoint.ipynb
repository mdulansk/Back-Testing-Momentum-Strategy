```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
```

```python
#Read in list of ticker symbols
tickers = pd.read_csv('./data/SP500_Symbols.csv', index_col= [0], header=0)
tickers.drop(['exchange_id','currency','created_date','last_updated_date'], axis=1, inplace=True)
tickers.head(10)
```

```python
stocks = pd.read_csv('./data/SP500_DailyPrices.csv', index_col= ['symbol_id','Date'], header=0, parse_dates=True)
```

```python
stocks.head()
```

### Create new column in stocks dataframe with ticker value from tickers df using id which is key
#### First set index in both dataframe to 'id' / 'symbol_id', then set 'ticker' column in stocks df equal to ticker column in tickers df.

```python
stocks.reset_index(inplace = True)
stocks.set_index('symbol_id', inplace = True)
```

```python
stocks.index
```

```python
tickers.index
```

```python
stocks['ticker'] = tickers['ticker']
```

```python
stocks.groupby('ticker')['Adj Close'].max() #Looks good so far.
```

```python
stocks.set_index(['ticker', 'Date'], inplace = True)
stocks.sort_index(inplace = True)
```

```python
stocks.loc[('AAPL','2000-01-03'), :]
```

```python
stocks.loc[('AAPL',['2000-01-03','2000-01-04']), :]
```

Note: When selecting a slice of values in a multiindex dataframe, you need to use the 'slice' method. You can't use :

```python
stocks.loc[('AAPL',[slice('2000-01-03','2000-01-06')]), :]
```
