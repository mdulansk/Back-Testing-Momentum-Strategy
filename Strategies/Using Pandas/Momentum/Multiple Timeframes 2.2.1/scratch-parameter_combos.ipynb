```python
# Get dataframe of all stats from previous backtests
all_stats_df = concat_stats_files() # This function returns stats from previous backtests
param_columns = [column for column in all_stats_df.columns if column in param_dict] # Get list of parameters we're testing
tested_combos_df = all_stats_df.loc[:,param_columns][0:5] # For our testing, we'll take first few rows for tested_combos_df
proposed_combos_df = all_stats_df.loc[:,param_columns][3:7] # For our testing, we'll take next few rows with some overlap for proposed_combos_df
tested_combos_df.head()
```

```python
# OLD WAY
# Create a dict of proposed parameter combo, which we can convert into df for comparison with
# Check to see if we've tested this combination of parameters before appending to combo test list
query_keys_values = [] # List to store proposed parameter combos to create query
combo_dict = {} # Create a dict of combo, value pairs to add to tested_combos_df
params = tested_combos_df.columns
for param in params:
    if param in param in param_dict: # Param exists in stats and params dict
        query_keys_values.append(f'{param}=={param_dict[param]}')
        combo_dict[param] = param_dict[param]
#query_str = 'rebal_int==3 and qty_long==43 and qty_med==20 and qty_short==10 and SMA_S==9 and SMA_L==139 and stop==0 and rolling_l==34 and rolling_m==16 and rolling_s==8'
query_str = ((' and ').join(query_keys_values)) # Create query string from our proposed parameter combos to query \
# tested_combos_df to see if we've tested before
print(f'query_str: {query_str}')
matched_rows = tested_combos_df.query(query_str)
print(f'matched_rows: {len(matched_rows)}\n')
if len(matched_rows) == 0: # If we don't find proposed combo in tested_combos_df, we'll add so we don't test again during this run
    #print(f'combo_dict: {combo_dict}')
    print(f'Param combo not used before, adding to combos to test:\n{combo_dict}')
    combos_df = pd.DataFrame(data=combo_dict, index=[0])
    tested_combos_df = pd.concat([tested_combos_df, combos_df], ignore_index=True).drop_duplicates(keep='first')
else:
    print(f'Param combo already tested:\n{combo_dict}')
```
