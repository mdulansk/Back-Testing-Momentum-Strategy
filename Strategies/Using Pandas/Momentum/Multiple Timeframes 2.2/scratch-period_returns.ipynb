```python
import pandas as pd
```

```python
return_csv_files = {'daily_returns_csv':'./cached_data/nasdaq_daily_log_returns.csv',
'weekly_returns_csv':'./cached_data/nasdaq_weekly_log_returns.csv',
'monthly_returns_csv':'./cached_data/nasdaq_monthly_log_returns.csv'}
tickers = ['NVDA', 'FTNT', 'MRVL', 'AMD', 'ZS']
data_start_date: str = '2010-01-01'
data_end_date: str = '2022-12-23'
```

```python
def read_data(returns_csv,tickers,data_start_date,data_end_date):
    returns_df = pd.read_csv(returns_csv, index_col=['Date'], parse_dates=True)
    selected_returns = returns_df.loc[data_start_date:data_end_date,tickers]
    print(f'Selected returns:\n{selected_returns.head(10)}')
    return(selected_returns)
```

```python
for name, returns_csv in return_csv_files.items():
    print(f'\n** {name} **')
    read_data(returns_csv,tickers,data_start_date,data_end_date)
```

```python
returns_csv = return_csv_files['weekly_returns_csv']
read_data(returns_csv,tickers,data_start_date,data_end_date)
```

```python
returns_csv = return_csv_files['monthly_returns_csv']
read_data(returns_csv,tickers,data_start_date,data_end_date)
```

```python
returns_csv = return_csv_files['weekly_returns_csv']
weekly_returns_df = read_data(returns_csv,tickers,'2010-02-07','2010-02-28')
#print(f'Selected returns:\n{weekly_returns_df.head()}')
```

```python
curr_rets = weekly_returns_df.mean(axis=1)
print(curr_rets)
```

```python
curr_tot_ret = weekly_returns_df.mean(axis=1).values.sum()
print(curr_tot_ret)
```
