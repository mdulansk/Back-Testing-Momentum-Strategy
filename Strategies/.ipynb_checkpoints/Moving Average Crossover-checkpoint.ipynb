### Moving Windows

```python
short_window = 40
long_window = 252
adj_close = aapl['Adj Close']
# Short moving window rolling mean
aapl['42'] = adj_close.rolling(window=short_window).mean()
# Long moving window rolling mean
aapl['252'] = adj_close.rolling(window=long_window).mean()
# Plot the adjusted closing price, the short and long windows of rolling means
aapl[['Adj Close', '42', '252']].plot(figsize=(22,10))

plt.show()
```

### Volatility Calculation

```python
# Define the minumum of periods to consider 
min_periods = 75 
daily_pct_change = adj_close.pct_change()
# Calculate the volatility
vol = daily_pct_change.rolling(min_periods).std() * np.sqrt(min_periods)
aapl['vol'] = daily_pct_change.rolling(min_periods).std() * np.sqrt(min_periods)
# Plot the volatility
#vol.plot(figsize=(22,10))
#plt.show()
```

## Building A Trading Strategy With Python

### Note: I'm not sure the positions logic is correct, since we should hold a position until we sell it, right? This situation may be handled in the backtest section below.

```python
# Initialize the short and long windows
short_window = 40
long_window = 100

# Initialize the `signals` DataFrame with the `signal` column
signals = pd.DataFrame(index=aapl.index) #Create new dataframe, using index from aapl df as index
signals.sort_index(inplace=True)
signals['signal'] = 0.0 #Create new column and set values to 0

# Create short simple moving average over the short window
signals['short_mavg'] = aapl['Close'].rolling(window=short_window, min_periods=1, center=False).mean() #Note the min_periods which means that the mean calc uses a much shorter period until it reaches the number of periods for the mean

# Create long simple moving average over the long window
signals['long_mavg'] = aapl['Close'].rolling(window=long_window, min_periods=1, center=False).mean()

# Set 'signls' value to 1 when short window avg crosses above long window. Set to -1 when opposite happens.
# It looks like we don't set a signal until we have enough data for the window, so it starts setting signal after the short_window value. That means the long_window avergage is not actually calculated with the correct # of period yet.
signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:] 
                                            > signals['long_mavg'][short_window:], 1.0, 0.0)   
#print(signals[:][short_window-5:])
# Generate trading orders
signals['positions'] = signals['signal'].diff() #'positions' value changes to 1 from 0 to indicate 'buy'
signals.fillna(0.0, inplace=True)
print(signals[:][short_window-5:]) #Print all columns and rows starting with index of short_window-5
```

#### Plot everything including buys and sells

```python
# Initialize the plot figure
fig = plt.figure(figsize=[28,14])

# Add a subplot and label for y-axis
ax1 = fig.add_subplot(111,  ylabel='Price in $') # Not sure what 111 is

# Plot the closing price
aapl['Close'].plot(ax=ax1, color='r', lw=2.)

# Plot the short and long moving averages
signals[['short_mavg', 'long_mavg']].plot(ax=ax1, lw=2.)

# Plot the buy signals
ax1.plot(signals.loc[signals.positions == 1.0].index, 
         signals.short_mavg[signals.positions == 1.0],
         '^', markersize=10, color='m')
         
# Plot the sell signals
ax1.plot(signals.loc[signals.positions == -1.0].index, 
         signals.short_mavg[signals.positions == -1.0],
         'v', markersize=10, color='k')
         
# Show the plot
plt.show()
```
